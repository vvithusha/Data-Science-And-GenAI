#### Real-world examples using lists in python

##### Example 01.manage A to Do List

```python
### simple to list in python
tasks = []

def show_tasks():
    if not tasks:
        print("no tasks in list")
    else:
        print("\n to-do list:")
        for i,task in enumerate(tasks, start=1):
            print(f"{i}.{task}")
            print()
def add_task(task):
    task = input("enter a new task:")
    tasks.append(task)
    print(f"task '{task}' added successfully!")

def remove_task():
    show_tasks()
    try:
        task_num = int(input("enter task number to remove:"))
        removed = tasks.pop(task_num - 1)
        print(f"task '{removed}' removed successfully!")
    except (IndexError, ValueError):
        print("invalid task number.")
def main():
    while True:
        print("\n to-do list manager")
        print("1.show tasks")
        print("2.add task")
        print("3.remove task")
        print("4.exit")
        choice = input("choose an option(1-4):")
        if choice == '1':
            show_tasks()
        elif choice == '2':
            add_task(input)
        elif choice == '3':
            remove_task()
        elif choice == '4':
            print("exiting the to-do list manager.")
            break
        else:
            print("invalid choice.please try again.")
if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox

# Main Window
root = tk.Tk()
root.title("To-Do List")
root.geometry("400x500")

tasks = []

# Frame for tasks
frame_tasks = tk.Frame(root)
frame_tasks.pack(pady=10)

# Scrollbar
scrollbar = tk.Scrollbar(frame_tasks)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

# Listbox with multiple selection
listbox = tk.Listbox(
    frame_tasks,
    width=40,
    height=15,
    yscrollcommand=scrollbar.set,
    selectmode=tk.MULTIPLE,
    font=("Arial", 12)
)
listbox.pack(side=tk.LEFT, fill=tk.BOTH)
scrollbar.config(command=listbox.yview)

# Entry box for new task
entry_task = tk.Entry(root, width=30, font=("Arial", 14))
entry_task.pack(pady=10)

# Functions
def add_task():
    task = entry_task.get()
    if task != "":
        listbox.insert(tk.END, task)
        entry_task.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task!")

def delete_task():
    selected = listbox.curselection()
    for index in selected[::-1]:
        listbox.delete(index)

def clear_all():
    listbox.delete(0, tk.END)

# Buttons
frame_buttons = tk.Frame(root)
frame_buttons.pack(pady=10)

btn_add = tk.Button(frame_buttons, text="Add Task", command=add_task, width=12)
btn_add.grid(row=0, column=0, padx=5)

btn_delete = tk.Button(frame_buttons, text="Delete Task", command=delete_task, width=12)
btn_delete.grid(row=0, column=1, padx=5)

btn_clear = tk.Button(frame_buttons, text="Clear All", command=clear_all, width=12)
btn_clear.grid(row=0, column=2, padx=5)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("To-Do List with Checkboxes")
root.geometry("400x500")

tasks = []  # store (task_text, IntVar for checkbox)

# Frame for tasks
frame_tasks = tk.Frame(root)
frame_tasks.pack(pady=10, fill="both", expand=True)

# Canvas + Scrollbar (so list can scroll if too long)
canvas = tk.Canvas(frame_tasks)
scrollbar = tk.Scrollbar(frame_tasks, orient="vertical", command=canvas.yview)
task_frame = tk.Frame(canvas)

task_frame.bind(
    "<Configure>",
    lambda e: canvas.configure(scrollregion=canvas.bbox("all"))
)

canvas.create_window((0, 0), window=task_frame, anchor="nw")
canvas.configure(yscrollcommand=scrollbar.set)

canvas.pack(side="left", fill="both", expand=True)
scrollbar.pack(side="right", fill="y")

# Entry box for new task
entry_task = tk.Entry(root, width=30, font=("Arial", 14))
entry_task.pack(pady=10)

# Functions
def add_task():
    task = entry_task.get()
    if task.strip() != "":
        var = tk.IntVar()
        cb = tk.Checkbutton(task_frame, text=task, variable=var, font=("Arial", 12))
        cb.pack(anchor="w", pady=2)
        tasks.append((task, var, cb))
        entry_task.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task!")

def delete_task():
    to_remove = [t for t in tasks if t[1].get() == 1]  # selected tasks
    for task, var, cb in to_remove:
        cb.destroy()
        tasks.remove((task, var, cb))

def clear_all():
    for task, var, cb in tasks:
        cb.destroy()
    tasks.clear()

# Buttons
frame_buttons = tk.Frame(root)
frame_buttons.pack(pady=10)

btn_add = tk.Button(frame_buttons, text="Add Task", command=add_task, width=12)
btn_add.grid(row=0, column=0, padx=5)

btn_delete = tk.Button(frame_buttons, text="Delete Selected", command=delete_task, width=12)
btn_delete.grid(row=0, column=1, padx=5)

btn_clear = tk.Button(frame_buttons, text="Clear All", command=clear_all, width=12)
btn_clear.grid(row=0, column=2, padx=5)

root.mainloop()
```
