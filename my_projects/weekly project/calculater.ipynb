```python
import math

def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "Error: Division by zero!"
    return a / b

def power(a, b):
    return a ** b

def sqrt(a):
    return math.sqrt(a)

def mod(a, b):
    return a % b

def trigo():
    angle = float(input("Enter the angle (in degrees): "))
    opp = input("Which operation (sin, cos, tan): ").lower()
    rad = math.radians(angle)  # convert to radians
    if opp == "sin":
        return math.sin(rad)
    elif opp == "cos":
        return math.cos(rad)
    elif opp == "tan":
        return math.tan(rad)
    else:
        return "Invalid choice! Please choose sin, cos, or tan."

while True:
    print("\nWelcome to the CLI Calculator!")
    choice = input("Enter operation (add, sub, mul, div, pow, sqrt, trigo, mod, help, exit): ").lower()

    if choice == "help":
        print("Available operations:")
        print("1. add\n2. sub\n3. mul\n4. div\n5. pow\n6. sqrt\n7. trigo\n8. mod\n9. exit")
        continue

    if choice == "exit":
        print("Exiting the calculator... Goodbye!")
        break

    # Operations that need two numbers
    if choice in ["add", "sub", "mul", "div", "pow", "mod"]:
        a = float(input("Enter first number: "))
        b = float(input("Enter second number: "))

        if choice == "add":
            print("Result:", add(a, b))
        elif choice == "sub":
            print("Result:", sub(a, b))
        elif choice == "mul":
            print("Result:", mul(a, b))
        elif choice == "div":
            print("Result:", div(a, b))
        elif choice == "pow":
            print("Result:", power(a, b))
        elif choice == "mod":
            print("Result:", mod(a, b))

    # Operations with one number
    elif choice == "sqrt":
        a = float(input("Enter a number: "))
        print("Result:", sqrt(a))

    elif choice == "trigo":
        print("Result:", trigo())

    else:
        print("Invalid operation! Type 'help' to see available commands.")
```
